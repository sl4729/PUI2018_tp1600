The homework below leverages data from NYC Open Data regarding Citywide Employee Payroll Information for base & OT salaries. The goal of the analysis was to read data from a .CSV file, manipulate it, and plot in a simple scatter plot with title and caption.

```python
#Download NY Citywide Payroll data from URL
#!curl https://data.cityofnewyork.us/api/views/k397-673e/rows.csv?accessType=DOWNLOAD > citywidepayroll.csv
```

```python
#read in data using pandas
pr=pd.read_csv("%s/citywidepayroll.csv"%os.getenv("PUIDATA"))
```

```python
pr.head()
```

```python
#Check the existing columns
pr.columns
```

```python
#drop all columns except Base Salary and OT hours
pr.drop([u'Regular Hours', u'Regular Gross Paid', u'Total OT Paid', u'Total Other Pay'], axis=1)
```

```python
import pylab as pl
%pylab inline
```

```python
#plot the OT hours vs Base Salary 
sc= pr.plot(x='Base Salary', y='OT Hours', kind='scatter')
sc.set_title("Citywide Workers OT Hours vs Base Salary")
```

Figure 1: The scatter plot above contains city employee data for the Fiscal Year 2017, with OT hours on the y axis and Base Salary in USD on the x axis. You can see that most OT hours are accrued by employees with lower salaries, including a particularly severe drop off as salaries near $150K.
